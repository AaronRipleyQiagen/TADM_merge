```python
lhpB_data = tadm_data.loc['NMDX_LHPB_Plasma_15_TADM']
lhpB_data['XPCR Module Serial'].unique()

import seaborn as sns
fig, axs = plt.subplots(4,4, figsize=(20,10),sharex=True, sharey=True)
plt.subplots_adjust(wspace=0.05)

cartridges = ['12370W', '12370X',  '12370Y', '12370Z']
xpcrs = ['V1086', 'V1292', 'V1293', 'V1296']
colors = {'12370W':'red', '12370X':'blue', '12370Y':'green', '12370Z':'orange'}
lhpB_data_plotable = lhpB_data.reset_index()
for item in lhpB_data_plotable.index:
    cartridge = lhpB_data_plotable.loc[item, 'Pcr Cartridge Lot']
    xpcr = lhpB_data_plotable.loc[item, 'XPCR Module Serial']

    pressure_values = lhpB_data_plotable.loc[item, [str(i) for i in range(0, 48510, 10)]].values

    sns.lineplot(x=[str(i) for i in range(0, 48510, 10)], y=pressure_values, color=colors[cartridge], ax=axs[xpcrs.index(xpcr), cartridges.index(cartridge)])


for cartridge in cartridges:
    axs[0, cartridges.index(cartridge)].set_title(cartridge)

for xpcr in xpcrs:
    axs[xpcrs.index(xpcr), 0].set_ylabel(xpcr)

for cartridge in cartridges:
    for xpcr in xpcrs:
        start, end = axs[xpcrs.index(xpcr),cartridges.index(cartridge)].get_xlim()
        axs[xpcrs.index(xpcr),cartridges.index(cartridge)].xaxis.set_ticks(np.arange(start, end, 1000))
    axs[xpcrs.index(xpcr),cartridges.index(cartridge)].set_ylabel('Time (ms)')

plt.show()
```
